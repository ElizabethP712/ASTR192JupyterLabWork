## Plotting Light Curves & Period Folding of TESS Data

```python
import lightkurve as lk
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from astropy.io import ascii
```

```python
data = ascii.read("http://tessebs.villanova.edu/")
print(data.keys())
print(data)
```

```python
print(data['TESS ID'])
print(data['P0 [days]'])
TESS_ID = data['TESS ID']
Periods = data['P0 [days]']
```

```python
Search_Indexes = np.where((Periods>8) & (Periods<12))
```

```python
#Find the light curves
#Limit search to SPOC is related to aperture photometry (think about the image w/ flux)
#Assign variables with time and flux
#Find the phase folding plots (w/ Period instead of time relative to flux)
```

```python
for p, _id in zip(Periods[Search_Indexes], TESS_ID[Search_Indexes]):
    print(p, _id)
    search_result = lk.search_lightcurve(f'TIC {_id}',author="SPOC")
    lc = search_result[0].download()
    time = lc.time.value
    flux = lc.flux.value
    magic_period = p 
    plt.figure(figsize=(10,5))
    plt.scatter(time%magic_period, flux, color='mediumpurple',s=0.1)
    plt.xlabel("Phase[days]")
    plt.ylabel("Flux")
    plt.title(f"Phase Folded Light Curve of TIC {_id}")
```

```python
for p, _id in zip(Periods[Search_Indexes], TESS_ID[Search_Indexes]):
    search_result = lk.search_lightcurve(f'TIC {_id}',author="SPOC")
    lc = search_result[0].download()
    time = lc.time.value
    flux = lc.flux.value
    magic_period = p 
    fig, ax = plt.subplots(nrows=1, ncols=2, sharex=False, sharey=False, figsize=(15,5))
    ax[0].scatter(time, flux, color='blue', s=0.1)
    ax[0].set_title(f"Observed Light Curve of TIC {_id}, Period {p}")
    ax[0].set_xlabel("Time [days]")
    ax[0].set_ylabel("Flux [e^-/s]")
    ax[1].scatter(time%magic_period, flux, color='red', s=0.1)
    ax[1].set_title(f"Phase Folded Light Curve of TIC {_id}, Period {p}")
    ax[1].set_xlabel("Phase [days]")
    ax[1].set_ylabel("Flux [e^-/s]")
    
#TIC 2372579 = The period seems off by quite a bit.
#TIC 4629065 = Equal mass/radius binary star system (b/c light dip is the same)
#
```
