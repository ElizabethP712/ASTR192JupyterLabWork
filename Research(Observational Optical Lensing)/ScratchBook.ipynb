## Imports

```python
from astropy.io import ascii
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Data

```python
data = ascii.read("https://github.com/AndyTza/PeriodicLSSTScaps/blob/main/data/eb_sample.ascii")
```

```python
time, flux = [], []

for val in data['col2'].data:
    sep = val.split(' ')
    time.append(float(sep[0]))
    flux.append(float(sep[1]))
```

```python
plt.scatter(time, flux, color='mediumpurple')
plt.xlabel("Time (in days)")
plt.ylabel("Flux")
```

```python
time, flux = np.array(time), np.array(flux)
```

```python
magic_period = 0.770579
plt.scatter(time%magic_period, flux, color='mediumpurple')
plt.xlabel("Time (in days)")
plt.ylabel("Flux")
```
