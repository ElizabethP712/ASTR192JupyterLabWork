```python
import lightkurve as lk
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from astropy.io import ascii
```

```python
from scipy.signal import savgol_filter
from scipy import interpolate

def running_median(x, y, bin_size=10): #Running Average Function
    """Return the runnign median of a given time series.
    
        Parameters:
        -----------
        x (array): time axis
        y (array): flux axis
        ww (float or int): bin length of the running median
        
        Rerturns:
        ---------
        X (array): running median time axis
        Y (array): running median flux axis
    
    """
    
        # Step calibrated specifically for TESS data
    bin_step = np.arange(x.min()-1, x.max()+3*bin_size, step=bin_size) #defining the intervals of seperation when finding average of flux; the step=bin_size is the width of the seperation, the smaller the gap the closer to the original function but that could cut out the lensing
   
    X, Y = [], [] # empty arrays
    for i in range(len(bin_step)-1): #This for loop goes through each time interval t1 and t2 and gives us all the flux values between that, and then it takes the median that is less sensitive to outliers
        try:
            wa = np.where((x>bin_step[i]) & (x<bin_step[i+1])) # going through each time bin inteval
            X.append(np.mean((bin_step[i+1], bin_step[i]))) # mean time 
            Y.append(np.nanmedian(y[wa])) # median flux 
        except:
            continue
    
    X, Y = np.array(X), np.array(Y)
    rmn = ~np.isnan(Y) # remove any remaining nans
    return X[rmn], Y[rmn]


def signal_finder(x_data, y_data, window=0.005, mode='dif', svg_filter_factor=10): #you give it light curve data; window is time interval
    """
        Signal finder retuns the residual of a smoothed interpolated light curve 
        subtracted by the interpolated raw light curve.
       
       Parameters:
       -----------
       x_data (array): time axis assuming a phase folded light curve
       y_data (array): normalized flux axis
       window (float): running median bin size in days
       mode (str): Currently supports diff which will return the time axis and residual or the interpolated arrays
       
       Returns:
       --------
       For mode: dif:
           time (array): average time between each detection
           residual (array): residual between the interpolated raw light curve and interpolated running median light curve
       
       For mode: diagnostic:
           x1 (array): interpolated time axis from the runnign median 
           f1(x1) (array):interpolated flux axis from the runnign median 
           
           x2 (array): interpolated time axis from the raw light curve 
           f2(x2) (array):interpolated flux axis from the raw light curve
    """
    
    N_interpolate = 10_000 # interpolation resolution, connects dots into what appears as a line
    
    # Average the raw data
    rm_x, rm_y = running_median(x_data, y_data,
                                bin_size=window) #running above function, and added window, defines running average
    
    # Linearly interpolate the running median  
    f1 = interpolate.interp1d(rm_x, rm_y, 
                              kind='linear') #connecting the dots for running medium
    
    # Linearly interpolate the raw data 
    f2 = interpolate.interp1d(x_data, y_data,
                              kind='linear') #connecting the dots for the original function
    
    # Create arrays
    x1 = np.linspace(rm_x.min(), rm_x.max(), N_interpolate) #interprelation needs a time axis, so f2 was the y-values and we need x-values connected to them for both curves.
    x2 = np.linspace(x_data.min(), x_data.max(), N_interpolate)
    
    
    if mode=='dif': 
        # Residual: interpolated raw data - running median interpolated data 
        df  =  f2(x2)-f1(x1) #the difference between the median curve and raw data curve
        yhat = savgol_filter(df, svg_filter_factor, 2) #Smooth the signal using a savgol filter using a 2nd order Polynomial
        return x1, yhat
    
    elif mode=='diagnostic':
        # Return the interpolated arrays for diagnotistc
        return x1, f1(x1), x2, f2(x2)
```

```python
#This is the interperlation for my chosen graphs.
InterestingTESS = np.array([355152971, 156325724, 140659980, 37248140])
#period = 6.003065 # days
data = ascii.read("EB_TESScat.ascii")
TESS_ID = data['TIC']
Periods = data['period']
for i in InterestingTESS:
    
    Search_Indexes = np.where((Periods>8) & (Periods<12) & (TESS_ID==i))
    for p, _id in zip(Periods[Search_Indexes], TESS_ID[Search_Indexes]):

        ### Light curve Processing ### 
        # Search for light curve for TIC 16703161 
        try:
            lkk = lk.search_lightcurve(f'TIC {_id}', author='SPOC', exptime=120)

            lc = lkk[0].download() # download the first available sector
            lc = lc.remove_nans() # remove nans from the light curve flux
            lc = lc.normalize() # normalize light curve

            # Define the time and flux variables
            time, flux = lc.time.value, lc.flux.value.unmasked
            phase = time%p # phase fold time 

            phase_sorter = np.argsort(phase) # sort phase

        except:
            continue

        # Final light curve product: 
        # t0: phase folded time & flux (NOTE: tts essentially is just making sure that the phase folded data are in proper order)
        t0, f0 = phase[phase_sorter], flux[phase_sorter] #the phase folded light curve's time and flux

        # Run the signal finding algorithm:
        # I will choose an average window size of 0.001 days
        # svg_filter_factor=150 (don't wory about this too much; we're applying this other filter to the data to make sure that it's overall smooth)

        se = signal_finder(t0, f0, window=0.001, svg_filter_factor=151) # returns phase and residual
        s0 = signal_finder(t0, f0, window=0.001, mode='diagnostic', svg_filter_factor=151) # returns the interpolated arrays! 

        ##### PLOTTING ######
        fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,4))

        ## PLOTTING THE INTERPOLATIONS ## 
        ax[0].plot(s0[2], s0[3], color='k', alpha=0.9, label='Raw Interpolated')
        ax[0].plot(s0[0], s0[1], color='hotpink', label='Running Average Interpolated')
        ax[0].grid(True, alpha=0.7)
        ax[0].set_title(f'TIC:{_id}')
        ax[0].set_ylabel("Flux")
        ax[0].set_xlabel("Phase [days]")
        ax[0].legend()

        ## PLOTTING THE RESIUAL PLOT ##
        ax[1].plot(se[0], se[1], lw=0.5)
        ax[1].set_ylim(0)
        ax[1].grid(True, alpha=0.7)
        ax[1].set_ylabel("Residual [ppm]")
        ax[1].set_xlabel("Phase [days]")
```
