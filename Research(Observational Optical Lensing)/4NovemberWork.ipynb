```python
search_result = lk.search_lightcurve('TIC 91961')
print(search_result)
# number (#) is the number of the visit Tess made to TIC 91961, mission is TESS(sectors refer to the name of the observing time interval, exposure times varied in seconds, don't worry about target_name as much, distance in arcsecs is a coordinate on the sky and 100 arcseconds would be farther away from the source, zero means it is spot on)
```

```python
search_result[0]
```

```python
lc = search_result[0].download()
```

```python
lc #Table takes the image w/ math and pixels and gives us information to make a light curve
```

```python
lc.plot()
```

```python
lc.time.value
```

```python
lc.flux.value #____ empty entries without measurements there
```

```python
time = lc.time.value
flux = lc.flux.value
magic_period = 5.282624782288264 
plt.scatter(time%magic_period, flux, color='mediumpurple',s=0.1)
plt.xlabel("Phase[days]")
plt.ylabel("Flux")
plt.title("Phase Folded Light Curve of TIC 91961")
```

```python
#In the graph above, the first dip is when a star is in front of the other and the second is when it is behind the other; there are some binary star systems where the stars are same mass which makes these graphs confusing, but for this one one of the stars is of lower mass and brightness which makes the dip less. 
```

```python
plt.scatter(time, flux, color='mediumpurple',s=0.1)
plt.xlabel("Time[days]")
plt.ylabel("Flux")
plt.title("Observed Light Curve")
```

```python
#In the graph above, we are seeing multiple periods of the orbit of the binary star system, TIC 91961, where the gap in data is when TESS was not observing the sky. 
```
